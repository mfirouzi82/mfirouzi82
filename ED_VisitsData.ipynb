## Data cleaning

```python
# Removing Nones 
m1 = wt['TWIB'].mean()
m2 = wt['AGE_NUM'].mean()
wt['TWIB'] = wt['TWIB'].fillna(m1)
wt['AGE_NUM'] = wt['AGE_NUM'].fillna(m2)
wt = wt.dropna(axis=0)

# Removing outliers in age column
wt.drop(wt[wt['AGE_NUM'] < 1].index, inplace = True)
wt.drop(wt[wt['AGE_NUM'] > 150].index, inplace = True)

# Removing outliers in TWIB column
q_low = wt["TWIB"].quantile(0.01)
q_hi  = wt["TWIB"].quantile(0.99)
wt = wt[(wt["TWIB"] < q_hi) & (wt["TWIB"] > q_low)]

# Converting columns TWIB and AGE_NUM to int64 dtype
wt = wt.astype({'AGE_NUM': int, 'TWIB': 'int'})
wt.head()
```

## Emergency Department visits change rates after Covid 19

```python
# Removing year from Date_of_visit column
edv_19['Date_of_visit'] = edv_19['Date_of_visit'].map(lambda x: x.lstrip('2019-'))
edv_20['Date_of_visit'] = edv_20['Date_of_visit'].map(lambda x: x.lstrip('2020-'))

# Changing index column
edv_19_ = edv_19.set_index('Date_of_visit')
edv_20_ = edv_20.set_index('Date_of_visit')

# Calculating change rates from 2019 to 2020
edv_20_19_ = edv_20_.subtract(edv_19_)  # Subtracting 2020 ED visits from 2019 ED visits
edv_20_19_ = edv_20_19_.div(edv_19_)    # Dividing (2020-2019) ED visits by 2019 ED visits

# Changing format to percentage
pd.options.display.float_format = '{:.2f}%'.format
edv_20_19_
```

## Physician initial assessment waiting time change after Covid 19

```python
# Changing index column
edia_19 = edia_19.set_index('Province/territory')
edia_20 = edia_20.set_index('Province/territory')

# Subtracting 2020 physician initial assessment waiting time from 2019 
edia_20_19 = edia_20.subtract(edia_19)  
edia_20_19
```

## Median waiting time for hospital bed change after Covid 19

```python
# Changing index column
edhb_19 = edhb_19.set_index('Province/territory')
edhb_20 = edhb_20.set_index('Province/territory')

# Subtracting 2020 median time waiting for hospital bed from 2019 
edhb_20_19 = edhb_20.subtract(edhb_19)  
edhb_20_19
```

# Visualization

## Line plot of moving average of total Emergency Department visits by day, 2019 vs 2020

```python
# Creating a subclass of ED visits datasets 

# Selecting columns
edv_19_4 = edv_19[['Date_of_visit','Total']]
edv_19_4 = edv_19_4.rename(columns = {'Total':'Total.2019'}, inplace = False)

# Removing '2019-' from Date_of_visit column
edv_19_4['Date_of_visit'] = edv_19_4['Date_of_visit'].map(lambda x: x.lstrip('2019-'))

# Adding moving average to the dataset
edv_19_4['MAVG_2019'] = edv_19_4.expanding(min_periods=7).mean()

# Selecting columns
edv_20_4 = edv_20[['Date_of_visit','Total']]
edv_20_4 = edv_20_4.rename(columns = {'Total':'Total.2020'}, inplace = False)

# Removing '2020-' from Date_of_visit column
edv_20_4['Date_of_visit'] = edv_20_4['Date_of_visit'].map(lambda x: x.lstrip('2020-'))

# Adding moving average to the dataset
edv_20_4['MAVG_2020'] = edv_20_4.expanding(min_periods=7).mean()

# Merging datasets                  
edv_19t20_t = pd.merge(edv_19_4, edv_20_4, on='Date_of_visit', how='outer')

# Changing Date_of_visit format
edv_19t20_t['Date_of_visit'] = edv_19t20_t['Date_of_visit'].replace({"3-": "March-", "4-": "April-", 
                             "5-": "May-", "6-": "June-"}, regex=True)

# Line plot of moving average of ED visits by day, comparing 2019 vs 2020
edv_19t20_t.plot(x='Date_of_visit', y=['MAVG_2019','MAVG_2020'], figsize=(18,10), fontsize=17)
plt.title('Line plot of moving average of 7 days of ED visits by day, comparing 2019 vs 2020', fontsize=20);
```

## Trend lines of Emergency Department visits March to June 2019 vs March to June 2020

```python
# Creating a subclass of ED visits datasets 

# Selecting columns
edv_19_3 = edv_19[['Date_of_visit','P.E.I.','N.S.','Que.','Ont.','Man.','Sask.','Alta.','B.C.']]
# Changing column names
edv_19_3 = edv_19_3.rename(columns = {'Ont.':'Ont.2019', 'Que.':'Que.2019','Alta.':'Alta.2019','B.C.':'B.C.2019',
                                     'Man.':'Man.2019', 'Sask.':'Sask.2019','N.S.':'N.S.2019','P.E.I.':'P.E.I.2019',}, 
                           inplace = False)

# Removing year from Date_of_visit column
edv_19_3['Date_of_visit'] = edv_19_3['Date_of_visit'].map(lambda x: x.lstrip('2019-'))

# Selecting columns
edv_20_3 = edv_20[['Date_of_visit','P.E.I.','N.S.','Que.','Ont.','Man.','Sask.','Alta.','B.C.']]
# Changing column names
edv_20_3 = edv_20_3.rename(columns = {'Ont.':'Ont.2020', 'Que.':'Que.2020','Alta.':'Alta.2020','B.C.':'B.C.2020',
                                     'Man.':'Man.2020', 'Sask.':'Sask.2020','N.S.':'N.S.2020','P.E.I.':'P.E.I.2020',}, 
                           inplace = False)

# Removing year from Date_of_visit column
edv_20_3['Date_of_visit'] = edv_20_3['Date_of_visit'].map(lambda x: x.lstrip('2020-'))

# Merging datasets 
edv_19t20_2 = pd.merge(edv_19_3, edv_20_3, on='Date_of_visit', how='outer')
edv_19t20_2

# Changing Date_of_visit format
edv_19t20_2['Date_of_visit'] = edv_19t20_2['Date_of_visit'].replace({"3-": "March-", "4-": "April-", 
                             "5-": "May-", "6-": "June-"}, regex=True)
edv_19t20_2
```

## Line plots of Emergency Department visits by day, comparing 2019 vs 2020 for provinces

```python
fig, axs = plt.subplots(4, 2)
fig.suptitle('Line plot of ED visits by day, comparing 2019 vs 2020', fontsize=30)
x = 'Date_of_visit'
edv_19t20_2.plot(x, y=['Ont.2019','Ont.2020'], title='Ontario', figsize=(24,32), fontsize=17, ax=axs[0, 0])
edv_19t20_2.plot(x, y=['Que.2019','Que.2020'], title='Quebec', figsize=(24,32), fontsize=17, ax=axs[0, 1])
edv_19t20_2.plot(x, y=['Alta.2019','Alta.2020'], title='Alberta', figsize=(24,32), fontsize=17, ax=axs[1, 0])
edv_19t20_2.plot(x, y=['B.C.2019','B.C.2020'], title='British columbia', figsize=(24,32), fontsize=17, ax=axs[1, 1])
edv_19t20_2.plot(x, y=['Man.2019','Man.2020'], title='Manitoba', figsize=(24,32), fontsize=17, ax=axs[2, 0])
edv_19t20_2.plot(x, y=['Sask.2019','Sask.2020'], title='Saskatchewan', figsize=(24,32), fontsize=17, ax=axs[2, 1])
edv_19t20_2.plot(x, y=['N.S.2019','N.S.2020'], title='Nova scotia', figsize=(24,32), fontsize=17, ax=axs[3, 0])
edv_19t20_2.plot(x, y=['P.E.I.2019','P.E.I.2020'], title='Prince Edward Island ', figsize=(24,32), fontsize=17, ax=axs[3, 1])

plt.show()
```

## Comparing median waiting time to physician initial assessment in Emergency Department
### March to June 2019 vs March to June 2020

```python
# Appending edia_19 and edia_20 dataframes

edia_19['Year'] = '2019'
edia_20['Year'] = '2020'
df1 = [edia_19, edia_20]
edia_1920 = pd.concat(df1)
edia_1920 = edia_1920.reset_index()
edia_1920
```

### Barplot of median waiting time to physician initial assessment in ED for March to June 2019 vs 2020

```python
fig, axs = plt.subplots(2, 2, figsize=(22,15))
fig.suptitle('Median time to physician initial assessment', fontsize=20)
sns.barplot(x="Province/territory", y="March", hue="Year", data=edia_1920, ax=axs[0, 0])
sns.barplot(x="Province/territory", y="April", hue="Year", data=edia_1920, ax=axs[0, 1])
sns.barplot(x="Province/territory", y="May", hue="Year", data=edia_1920, ax=axs[1, 0])
sns.barplot(x="Province/territory", y="June", hue="Year", data=edia_1920, ax=axs[1, 1])
plt.show()
```
