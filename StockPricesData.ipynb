## Converting Date format to mm/dd/yyyy 

```python
def date_fr(df):
    df['Date'] = pd.to_datetime(df.Date)
    df['Date'] = df['Date'].dt.strftime('%m/%d/%Y') 
    print(df.head())
```

## Correcting years between 1970 to 1999

```python
def year_adj(df):
    df['year'] = pd.to_datetime(df['Date'], errors='coerce').dt.year
    df['year'] = np.where((df['year'] > 2069) & (df['year'] <= 2099), df['year'] - 100 , df['year'])
    df["Date"] = df["Date"].astype(str)
    df["year"] = df["year"].astype(str)
    df['Date'] = df.Date.str[0:5] + '/' + df["year"]
    df["year"] = df["year"].astype(int)
    print(df.head())
```

```python
# IBM dataframe
year_adj(IBM)
```

```python
# GE dataframe
year_adj(GE)
```

```python
# ProcterGamble dataframe
year_adj(ProcterGamble)
```

```python
# CocaCola dataframe
year_adj(CocaCola)
```

```python
# Boeing dataframe
year_adj(Boeing)
```

# Visualization

```python
# Line plot of price and 7 days moving average of Coca-Cola stock price

CocaCola1 = CocaCola.drop(['year'], axis=1)
CocaCola1['MAVG'] = CocaCola1.expanding(min_periods=7).mean()
CocaCola1.plot(x='Date', y=['StockPrice','MAVG'], figsize=(18,6), fontsize=17)
plt.title('Line plot of price and 7 days moving average of Coca-Cola stock price', fontsize=20);
```

```python
#1. Identify the year during which Coca-Cola had the highest/lowest stock price?

sp_max = CocaCola['StockPrice'].max()
y_max = CocaCola[CocaCola['StockPrice'] == sp_max]
y_max = y_max.values.tolist()
print('Year of highest price: ', y_max[0][2])

sp_min = CocaCola['StockPrice'].min()
y_min = CocaCola[CocaCola['StockPrice'] == sp_min]
y_min = y_min.values.tolist()
print('Year of lowest price: ', y_min[0][2])
```

```python
#2. What calendar year did it look to have the biggest (Year-over-Year) percentage increase?

# Groupby on year column by mean
CocaCola2 = CocaCola.drop('Date', axis=1)
CocaCola2 = CocaCola.groupby('year').mean()

# Finding percentage change rate 
CocaCola2['change'] = CocaCola2.pct_change(fill_method ='ffill')
ch_max = CocaCola2['change'].max()
yc_max = CocaCola2[CocaCola2['change'] == ch_max]
yc_max.reset_index(inplace=True)
yc_max = yc_max.values.tolist()
print('Year of highest percentage increase: ', int(yc_max[0][0]))
```

## Monthly Trend Analysis

```python
#1. For IBM, compare the average stock price for each month to the its overall average stock price
# and identify all the months for which IBM historically had a higher stock price (we call this overindexing)?
# Which month over-indexed the most? 

IBM5 = IBM.copy()
mean = IBM5['StockPrice'].mean()
IBM5['month'] = pd.to_datetime(IBM5['Date'], errors='coerce').dt.month
IBM5 = IBM5.drop(['year'], axis=1)
IBM5 = IBM5.groupby('month').mean()
IBM5['Mean'] = mean
IBM5['Over_index'] = np.where((IBM5['StockPrice'] > mean) , 'Yes' , 'No')
IBM5
```

```python
#2. Repeat the tapply() function you used to solve the last question for each of the 4 remaining
# companies. Do any of two or more companies have their highest stock price in the same months
# as each other? Which companies and months does this happen for?

GE5 = GE.copy()
CocaCola8 = CocaCola.copy()
ProcterGamble5 = ProcterGamble.copy()
Boeing6 = Boeing.copy()

def over_idx(df):
    mean = df['StockPrice'].mean()
    df['month'] = pd.to_datetime(df['Date'], errors='coerce').dt.month
    df = df.drop(['year'], axis=1)
    df = df.groupby('month').mean()
    df['Mean'] = mean
    df['Over_index'] = np.where((df['StockPrice'] > mean) , 'Yes' , 'No')
    return df

print('GE: \n', over_idx(GE5))
print()
print('CocaCola: \n', over_idx(CocaCola8))
print()
print('P&G: \n', over_idx(ProcterGamble5))
print()
print('Boeing: \n', over_idx(Boeing6))

# IBM, GE and CocaCola are almost similar in months with the highest stock price in the first months of year
```
